1. Let $A,B$ be two convex sets. Define $C:= A+B = \left\{ \boldsymbol{a} + \boldsymbol{b} |~ \boldsymbol{a} \in A,~ \boldsymbol{b} \in B \right\}$. Show that $C$ is convex.
2. Let $A_k$, $k=1,2,\cdots,K$, be convex sets. Show that $A := \bigcap_{k=1}^K A_k$ is convex. 
3. Show that a set is convex if and only if its intersection with any line is convex. 


1.
$$
c_1 = (a_1 + b_1), \; c_2 = (a_2+b_2) \\
\lambda (a_1 + b_1) + (1-\lambda)(a_2+b_2) = \lambda a_1 + (1-\lambda)a_2 + \lambda b_1 + (1-\lambda)b_2 \\
\text{If } a = \lambda a_1 + (1-\lambda)a_2, \; a \in A \text{ by definition of convex set} \\
\text{If } b = \lambda b_1 + (1-\lambda)b_2, \; b \in B \text{ by definition of convex set} \\
c = \lambda c_1 + (1-\lambda)c_2 = a + b, \; \lambda \in [0,1] \\
\text{Thus } c \in C \text{ by definition of } C \text{ so } C \text{ is convex}\\
$$
2.
For any points $x$ and $y$ in $A_1\bigcap A_2$, $x$ and $y$ must be in $A_1$ and $A_2$ by definition. Also any linear combination of $x$ and $y$, $p = \lambda x + (1-\lambda)y, \; \lambda \in [0,1]$, $p$ must be in $A_1$ and $A_2$ as both sets are convex. Since $p$ is in $A_1$ and $A_2$ it must be in $A_1\bigcap A_2$ by definition of the union. Thus if $A_1$ and $A_2$ are convex, $A_1\bigcap A_2$ must also be convex. This logic can be used iteratively with $A := \bigcap_{k=1}^K A_k$.

3. 
Take a line $L$ such that the intersection of $L$ and the set $S$ isn't convex. $p_1$ and $p_2$ are the furthest points in this intersection such that:
$$
p_1,p_2 = \text{argmax}_{p_1,p_2 \in L\bigcap S}(|p_2| - |p_1|)
$$
A point $p$ on $L$ can be described with the following equation due to the definition of a line
$$
p = p_1 + a(p_2-p_1), a \in \R \\
p = ap_2 + (1-a)p_1 \\
$$
if $p$ is in the intersection of the line and set we have 
$$
|p_1| \le |p| \le |p_2| \text{ by the definition of } p_1 \text{ and } p_2 \\
\text{Therefore} \\
p = ap_1 + a(p_2-p_1), a \in [0,1]\\
$$
If the intersection is not convex there exists a $p$ such that $p = ap_1 + a(p_2-p_1), a \in [0,1]$, which is not in the union of the two sets, as if for all $p$ that meet this condition are in the union the union in convex. We know $p$ must be on the line $L$ by definition and so must not be in the set $S$. Therefore by the definition of a convex set $S$ is not convex. Therefore is there exists a line for which $L\bigcap S$ is not convex $S$ must also not be convex and if S is not convex there will exists a line that makes these conditions true.

If we consider the 2D wavelet transform $\mathcal{W}(X)$ as being a linear operator on $X$ in the form $WXW^T$ we could find the proximal operator. However, if we consider instead a vectorised image instead of an image we can use the kronecker product form to give $\mathcal{W}(x) = (W \otimes W^T)x$ where $x = vec(X)$. We can write $W \otimes W^T = T$ and solve for the proximal operator of $g(x) = \lambda\|Tx\|$. 

$$
\text{Prox}_{\lambda,g}(z)_{(i)} = T^T(\text{sign}(x_{(i)})  \text{max}(|x_{(i)}| - \lambda, 0))

As previously we could consider the sampling a radom set of rows and columns of the image allowing us to write $\mathcal{P}_{\Omega}(\mathcal{F}(X)) = PFXF^TP^T$. With this identity we are able to differentiate $f(x)$.
 
$$
%  \mathcal{P}_{\Omega} ( \mathcal{F}
\mathcal{F}(X) = FxF^T \\
\mathcal{P}_{\Omega} = PxP^T \\
f(x) = \frac{1}{2}\|y - PFxF^TP^T\|_2^2 \\
A = PF \\
f(x) = \frac{1}{2}\|y - AxA^T\|_2^2 \\
\nabla f(x) = A^T \otimes A (AxA^T - y) \\
$$

However, in order to do some more complex sampling of the image it may be easier to use the vectorised version of the image an construct a matrix that can multiply this vector to give the new image. 

$$
x = \text{vec}(X) \\
M = PF \\
vec(MXM^T) = M \otimes M^T vec(X) = M \otimes M^T x \\
\text{We consider the function } \\
f(x) = \frac{1}{2}\|y - M \otimes M^T x\|_2^2 \\
A = M \otimes M^T = (PF)  \otimes (PF)^T \\
f(x) = \frac{1}{2}\|y - A x\|_2^2 \\
\nabla f(x) = A^T(A x - y)
$$

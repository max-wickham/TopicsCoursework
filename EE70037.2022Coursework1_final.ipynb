# 3 Adjoint (10)

Let 
$$
\begin{align*}
    \mathsf{A}:~ \mathbb{R}^n 
    & \rightarrow \mathbb{R}^m \\ 
    \boldsymbol{x}
    & \mapsto \boldsymbol{y} = \boldsymbol{A} \boldsymbol{x}.
\end{align*}
$$

The adjoint of the linear operator $\mathsf{A}$, denoted by $\mathsf{A}^*$, is defined via 
$$
\begin{align*}
    \left\langle \mathsf{A}(\boldsymbol{x}),\boldsymbol{y} \right\rangle
    & = \left\langle \boldsymbol{x},\mathsf{A}^*(\boldsymbol{y}) \right\rangle
\end{align*}
$$
for all $\boldsymbol{x} \in \mathbb{R}^n$ and $\boldsymbol{y} \in \mathbb{R}^m$. 

It is straightforward to verify that 
$$
\begin{align*}
    & \mathsf{A}^{*}(\boldsymbol{y}) = \boldsymbol{A}^{\mathsf{T}} \boldsymbol{y}.
\end{align*}
$$



**Remark** If $\mathsf{A}:~\mathbb{C}^n \rightarrow \mathbb{C}^m$, then $\mathsf{A}$ is generally represented by a matrix $\boldsymbol{A} \in \mathbb{C}^{m \times n}$. The adjoint of $\mathsf{A}$, i.e., $\mathsf{A}^*$, can be represented by $\boldsymbol{A}^{\mathsf{H}}$, where the superscript $\mathsf{H}$ denotes `conjugate transpose'. 


##  3.1  

- Consider the linear map given by 
    $$
    \begin{align*}
        & \boldsymbol{Y} = \mathsf{C}(\boldsymbol{X}) = \boldsymbol{A} \boldsymbol{X} \boldsymbol{B}
    \end{align*}
    $$
    C*(y) = AtyBt

    - Find the adjoint of $\mathsf{C}$ and represent it using matrices $\boldsymbol{A}$ and $\boldsymbol{B}$.
    - Let $\boldsymbol{Z} = \mathsf{C}^*( \mathsf{C} (\boldsymbol{X}) )$. 
       - Find the closed form to compute $\boldsymbol{Z}$ using $\boldsymbol{A}$, $\boldsymbol{B}$, and Kronecker product.
       - Find the closed form to compute $\boldsymbol{Z}$ using $\boldsymbol{A}$ and $\boldsymbol{B}$ but not using Kronecker product.
       - Compare the above two approaches in terms of time complexity (only consider number of multiplications) and space complexity.

##   3.1 Answer 
Adjoint of C:
$$
\langle C(X), Y \rangle = \langle X , C^*(Y) \rangle \\
\langle AXB, Y \rangle = \langle X , C^*(Y) \rangle \\
\text{trace}((AXB)^TY) = \text{trace}(X^TC^*(Y)) \\
\text{trace}(B^TX^TA^TY) = \text{trace}(X^TC^*(Y)) \\
\text{Since by the cyclic property of the trace }
\text{trace}(B^TX^TA^TY) = \text{trace}(X^TA^TYB^T) \\
C^*(Y) = A^TXB^T \text{ fits the definition} \\
$$
Finding Z:
$$
C^{*}(C(X)) = C^{*}(Y) \\
\text{For the mixed-product property of the Kronecker product:} \\
vec(C^{*}(Y)) = vec(A^TYB^T)
= (B \oplus A^T)vec(Y) \\
\text{substituting Y and using the same identity again on X:}\\
vec(C^{*}(Y)) = (B \oplus A^T)vec(AXB) = (B \oplus A^T)(B^T \oplus A)vec(X)\\ 
\text{Thus,}\\
Z = vec^{-1}((B \oplus A^T)(B^T \oplus A)vec(X)) = vec^{-1}((BB^T \oplus AA^T)vec(X))
$$
or by substitution of C and C*:
$$
Z = A^TAXBB^T
$$


 
- closed form solution with kronecker product: $ vect^{-1}((m\times m \oplus m \times m)(nn \times 1)) = vect^{-1}((mm \times mm)(nn \times 1))$
- closed form solution without kronecker: 4 multiplications, $Z \in R^{n \times n}$ result

The Kronecker product approach has higher time and space complexity. The Kronecker method includes two matrix multiplications and one Kronecker multiplication. The Kronecker multiplication has a $O(abpq)$ for $a\times b$ and $p\times q$ matrices, which gives $O(n^4)$, these multiplication all have to be stored giving the same spacial complexity. In the none Kronecker if we say a matrix multiplication has $O(n^2)$ complexity and we have 4 matrix multiplications we have $O(n^2)$ times and space complexity. The space complexity is the same as the time as at each stage of multiplication we just have to store the result of the previous multiplication.


## 4.1 Recursion

```julia
# Recursion
function fib1(k)
    if k <= 2
        return 1
    end
    return fib1(k-2) + fib1(k-1)
end
```


##  4.2 Memoization

```julia
# Memoization
function fib2(k)
    fibv = zeros(BigInt,k);
    fibv[1] = 1; 
    fibv[2] = 1; 
    for i in 3:k
        fibv[i] = fibv[i-1] + fibv[i-2];
    end 
    return fibv[k];
end 

```


##  4.3 Matrix-vector multiplication

 Based on that 
    $$
    \begin{align*}
        & \begin{bmatrix}

        \end{bmatrix}
        = \begin{bmatrix}
            1 & 1 \\

        \end{bmatrix}^k

```julia
# Matrix-Vector Multiplication
function fib3(k)
    if k <= 1
        return 1
    end

    k -= 2
    A = big.([ 1 1; 1 0 ]);
    fibv0 = [1; 1];
    fibvk = A^k*fibv0;
    return fibvk[1];
end
```


##  4.4 Matrix decomposition

 Using eigen-decomposition of the matrix $\boldsymbol{A}$, we can speed up the generation of Fibonacci series further. 

   Let $\boldsymbol{\Lambda} = [\lambda_1,\lambda_2]^{\mathsf{T}}$ be the eigenvalue vector, and $\boldsymbol{V} = [\boldsymbol{v}_1,\boldsymbol{v}_2]$ be the corresponding eigenvector matrix. Write $\boldsymbol{A}^k$ in terms of $\boldsymbol{\Lambda}$ and $\boldsymbol{V}$. 
  
  Complete the following functions:

```julia
# Matrix Decomposition
function fib4(k)
    if k <= 1
        return 1
    end
    k -= 2;
    A = [ 1 1; 1 0 ];
    fibv0 = [1; 1];
    Lambda, V = eigen(A); 
    Lambda[1] = Lambda[1]^k;
    Lambda[2] = Lambda[2]^k;
    Λ = diagm(Lambda);
    Q = V;
    # fill the contents here 
    Ak = Q*Λ*inv(Q);
    fibvk = Ak*fibv0;

    return BigInt(round(fibvk[1]));
end 
```

##   7.1 Answer 

Given a fixed parameter $\gamma$ and vector $\bold{z} = (z_{im1}, ..., z_{imn})^T$ with $|z_{i1}|>|z_{i2}|>...>|z_{in}|$, 
we define $N_m =\{ \bold{x} \in R^n | \| \bm{x} \|_0$ = m \}, i.e. the set of orthogonal vectors that have m total non-zero entries ($N_m$ disjoint partition of $R^n$).
Then:
$$
\begin{align*}
    \mathcal{M}_{\| \cdot \|_0}(\bold{z}) = \min_x~ \| x \|_0 + \frac{1}{2 \gamma}(\bold{x}-\bold{z})^2  
\end{align*}
$$
with $x \in N_m$. Now, considering $\min_x~ \| x \|_0 = m = i$ we can rewrite as:
$$
\begin{align*}
    \mathcal{M}_{\| \cdot \|_0}(\bold{z}) = i  + \min_x~ \frac{1}{2 \gamma} (\bold{x}-\bold{z})^2 
\end{align*}
$$
with $x \in N_i$. To minimise $\min_x~ \frac{1}{2 \gamma}(\bold{x}-\bold{z})^2$, we have to choose x so that it cancels out the first n entries with largest magnitude in z, i.e. the top $m_1$ to $m_i$ entries, given that z is ordered and i are the max number of entries we can cancel out. Therefore we get:
$$
\begin{align*}
     x_j = z_{nj}\text{ if } j\leq{n-i} \text{ and } x_j = 0 \text{ otherwise}
\end{align*}
$$
and 
$$
\begin{align*}
    \mathcal{M}_{\| \cdot \|_0}(\bold{z})  = i+ \frac{1}{2 \gamma} \sum_{j=n-i+1} (z_{mj})^2
\end{align*}
$$
Now, to find the proximal operator, corresponding to the $\arg \min \mathcal{M}_{\| \cdot \|_0}(\bold{z})$, e.g. the value that minimises the Moreau envelop, we have to consider the trade-off between choosing an i that minimises the linear term in i but maximises he number of entries we can cancel out with x (i term in the summation). Given that $i = m = \min_x~ \| x \|_0$ can be expressed as a sum of 1s, we have to find the first term $z_{mj}$ that is greater than 1, meaning that we have passed the threshold beyond which decreasing i does not minimise the equation anymore. This threshold can be expressed as:
$$
\begin{align*}
    (z_{mj})^2 /(2 \gamma) > 1 \\
    (z_{mj})^2 > 2 \gamma
\end{align*}
$$
Which expresses the square relationship between the parameter $\gamma$ and the threshold of minimisation corresponding to the proximal operator.



# 8 Compressed Sensing of Multi-tone Audio Signal (25)

The DFT of $\boldsymbol{x} \in \mathbb{C}^N$ can be defined as 
$$
\begin{align*}
    & \boldsymbol{X} 
    = \mathsf{F}(\boldsymbol{x}) 
    = \boldsymbol{F}(\boldsymbol{x})
    \in \mathbb{C}^N,
\end{align*}
$$

where
$$ 
\begin{align*}
    & X[k] = \sum_{n=1}^{N} 
     x_n\exp\left\{-2\pi j (n-1)(k-1)/N\right\},
\end{align*}
$$
the matrix $\boldsymbol{F}$ is given by 
$$
\begin{align*}
    & \boldsymbol{F}_{p,q} =\exp\left\{-j \frac{2\pi}{N} (p-1)(q-1)\right\},
\end{align*}
$$
and $\boldsymbol{F}_{p,q}$ denotes the element in the $p$-th row and $q$-th column of $\boldsymbol{F}$. 

The inverse DFT of $\boldsymbol{X} \in \mathbb{C}^N$ is therefore defined as
$$ 
\begin{align*}
    & \boldsymbol{x} 
    = \mathsf{F}^{-1} (\boldsymbol{X}) 
    = \boldsymbol{F}^{-1}(\boldsymbol{X})
    \in \mathbb{C}^N,
\end{align*} 
$$
where the matrix $\boldsymbol{F}^{-1}$ is explicitly given by 
$$
\begin{align*}
    & (\boldsymbol{F}^{-1})_{p,q} = \frac{1}{N} \exp\left\{j \frac{2\pi}{N} (p-1)(q-1)\right\}.
\end{align*}
$$

```julia
function dft_matrix(N::Int64)
    dftM = Array{ComplexF64,2}(undef,N,N);
    dftMinv = copy(dftM); 
        
    for q in 1:N 
        for p in 1:N
            dftM[p,q] = exp(
                (-1im)*((2*π)/N)*(p-1)*(q-1)
                );
            dftMinv[p,q] = exp(
                (1im)*((2*π)/N)*(p-1)*(q-1)
                );
        end
    end
    dftMinv = dftMinv/N; 

    

    return dftM,dftMinv
end

# Test dft_matrix and fft 
# N = 4096; 
# dftM,dftMinv = dft_matrix(N);

# x = randn(ComplexF64,N); 
# println("test for correctness");
# @show norm( dftM*x - fft(x) );
# @show norm( dftMinv*x - ifft(x) );

# println("runing time comparison");
# @btime dftM * x; 
# @btime fft(x); 
```

### 8.3.2 Implement SP to Solve Sparse Recovery (5)





```julia
function SupH(s, a)
    # Hard thresholding function
    x = zeros(ComplexF64, length(a));
    result = Set();
    for _ in 1:s
        maxval, index = findmax(abs.(a-x));
        x[index] = a[index];
        push!(result,index)
    end
    return result;
end

function SP(S, L, y)
    t = SupH(S,transpose(L)*y);
    yr = y
    l = 0
    b = zeros(ComplexF64, N);
    X_hat = zeros(ComplexF64, N);
    a_cross = pinv(L);
    a_cross_T_y = a_cross*y;
    while l < S
        l += 1
        union!(t,SupH(S,transpose(L)*y))
        for index in 1:N
            if index in t
                b[index] = a_cross_T_y[index];
            else
                b[index] = 0;
            end
        end
        t = SupH(S,b)
        for index in 1:N
            if index in t
                println(a_cross_T_y[index])
                X_hat[index] = (N/length(y))*a_cross_T_y[index];
            else
                X_hat[index] = 0;
            end
        end
        yr = y - L*X_hat;
    end
    return real((dftMinv*X_hat))
end

x_hat = SP(6,L,y);

fig_x = Figure();
ax_x = Axis(fig_x[0, 1],
    title = "SP Time Signal",
    xlabel = "n",
    ylabel = "x[n]"
);
lines!(ax_x, collect(0:N-1), x);
lines!(ax_x, collect(0:N-1), x_hat);
fig_x
```

### 8.3.3 Implement OMP to Solve Sparse Recovery (5)

```julia
function OMP(S, L, y)
    l = 1
    T = Set{Int64}()
    X_hat = zeros(ComplexF64, N);
    yr = y
    A_cross = pinv(L);
    while l < S
        i_max = 0;
        final_index = 0;
        for index in 1:N
            i = abs(dot(L[:,index],yr))
            if i > i_max
                i_max = i;
                final_index = index;
            end
        end
        push!(T,final_index);
        X_hat[final_index] = (N/length(y))*(A_cross*y)[final_index];
        yr = y - L*X_hat;
        l = l+1;
    end
    return real((dftMinv*X_hat))
end

x_hat = OMP(6,L,y);

fig_x = Figure();
ax_x = Axis(fig_x[0, 1],
    title = "OPM Time Signal",
    xlabel = "n",
    ylabel = "x[n]"
);
lines!(ax_x, collect(0:N-1), x);
lines!(ax_x, collect(0:N-1), x_hat);
fig_x
```
